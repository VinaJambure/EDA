```python
def bar_plot(variable):
    var = df[variable] 
    varValue = var.value_counts()  
    plt.figure(figsize=(5,3))
    plt.bar(varValue.index.astype(str), varValue)  
    plt.xticks(rotation=45)  
    plt.ylabel('StudentID')
    plt.xlabel(variable)
    plt.title(f'Distribution of {variable}')
    plt.show()
```

```python
columns_plot = ['Workshops/Certifications', 'AptitudeTestScore', 'SoftSkillsRating'] 
for i in columns_plot:
    bar_plot(i)
```

```python
import matplotlib.pyplot as plt

def hist_plot(vari):
    var = df[vari]  
    plt.figure(figsize=(9,3))
    plt.hist(var, bins=10, edgecolor='black', alpha=0.7)  
    plt.xlabel(vari)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {vari}')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
    print(f'{vari}:\n{var.describe()}')  
```

```python
numeric_columns = ['CGPA', 'Internships', 'Projects', 'AptitudeTestScore', 'SoftSkillsRating', 'SSC_Marks', 'HSC_Marks']
for col in numeric_columns:
    hist_plot(col)
```

```python
def box_plot(vari, df):
    plt.figure(figsize=(9, 4))
    sns.boxplot(y=df[vari])
    plt.title(f'Box Plot of {vari}')
    plt.ylabel(vari)
    plt.show()
```

```python
colum_plot=['CGPA', 'Internships', 'Projects', 'Workshops/Certifications','AptitudeTestScore', 'SoftSkillsRating',]
for i in colum_plot:
    if df[i].dtype in ['int64', 'float64']:  
        box_plot(i, df)
```
